```python
import numpy as np
import pandas as pd
import glob
import os
print(os.getcwd())
import re
import nltk
from sklearn.feature_extraction.text import CountVectorizer
```

```python
#base de datos de prueba
data = pd.read_csv("../datos/base_apariencia.csv", encoding="latin1")
```

```python
#Para ver un ejemplo de tweet
data.loc[150,['text']]
```

```python
csv_files = glob.glob('../datos/*.csv')
dfs = []
for filename in csv_files:
    dfs.append(pd.read_csv(filename, encoding = "latin1"))

# Concatenate all data into one DataFrame
data = pd.concat(dfs, ignore_index=True)
```

```python
base_analisis = data.loc[:,["text","tipo","motivo"]]
base_analisis.tail()
```

```python
# Remove punctuation
base_analisis['text'] = base_analisis['text'].map(lambda x: re.sub('[,\.!?]', '', x))
# Convert the titles to lowercase
base_analisis['text'] = base_analisis['text'].map(lambda x: x.lower())
# Print out the first rows of papers
base_analisis['text'].head()
```

```python
#nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords

stop_words_sp = set(stopwords.words('spanish'))
```

```python
stop_words_sp = set(stopwords.words('spanish'))
stop_words_en = set(stopwords.words('english'))
```

```python
stop_words = stop_words_sp | stop_words_en
from nltk import tokenize
```

```python
tweets = base_analisis['text']
```

```python
# Remover URLs, RTs, y twitter handles
for i in range(len(base_analisis['text'])):
    base_analisis['text'][i] = " ".join([word for word in base_analisis['text'][i].split()
              if 'http' not in word and '@' not in word and '<' not in word and 'RT' not in word])
#Monitorear que se removieron las menciones y URLs
print("------Después de remover menciones y URLs --------")
print(base_analisis['text'][100])
#Remover puntuación, se agregan símbolos del español
base_analisis['text'] = base_analisis['text'].apply(lambda x: re.sub('[¡!@#$:).;,¿?&]', '', x.lower()))
base_analisis['text'] = base_analisis['text'].apply(lambda x: re.sub('  ', ' ', x))
#Monitorear que se removió la puntuación y queda en minúsculas
print("------Después de remover signos de puntuación y pasar a minúsculas--------")
print(base_analisis['text'][100])
```

```python
# Import the wordcloud library
from wordcloud import WordCloud
# Join the different processed titles together.
long_string = ','.join(list(base_analisis['text'].values))
# Create a WordCloud object
wordcloud = WordCloud(background_color="white", max_words=5000, contour_width=3, contour_color='steelblue')
# Generate a word cloud
wordcloud.generate(long_string)
# Visualize the word cloud
wordcloud.to_image()
```

```python
count_vectorizer = CountVectorizer(stop_words=stop_words)
```

```python
count_data = count_vectorizer.fit_transform(base_analisis['text'])
```

```python
count_data.shape
```

```python
# Load the LDA model from sk-learn
from sklearn.decomposition import LatentDirichletAllocation as LDA
 
# Helper function
def print_topics(model, count_vectorizer, n_top_words):
    words = count_vectorizer.get_feature_names()
    for topic_idx, topic in enumerate(model.components_):
        print("\nTopic #%d:" % topic_idx)
        print(" ".join([words[i]
                        for i in topic.argsort()[:-n_top_words - 1:-1]]))
        
# Tweak the two parameters below
number_topics = 7
number_words = 20
# Create and fit the LDA model
lda = LDA(n_components=number_topics, n_jobs=-1)
lda.fit(count_data)
# Print the topics found by the LDA model
print("Topics found via LDA:")
print_topics(lda, count_vectorizer, number_words)
```
