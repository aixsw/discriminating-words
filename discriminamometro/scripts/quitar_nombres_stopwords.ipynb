```python
import pandas as pd
```

```python
nombres=['apellidos.csv','apellidos-20.csv','hombres.csv','mujeres.csv']
```

```python
for item in nombres:
    tabla = pd.read_csv(item)
    tabla['nombre']
    for i in range(0,len(tabla['nombre'])):
        stop.add(str(tabla['nombre'].loc[i]).lower())
        
```

```python
stop
```

```python
pickleFile = open('StopWords.p', 'wb')
pickle.dump(stop, pickleFile)
pickleFile.close()
```
